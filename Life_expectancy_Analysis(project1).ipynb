```python
correlation_matrix = LXA2[['Life expectancy ', 'Adult Mortality', ' BMI ', ' HIV/AIDS']].corr()
correlation_matrix
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = LXA2[['Life expectancy ', 'Hepatitis B', 'Polio']].corr()
print(correlation_matrix)

plt.figure(figsize=(5,5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix: Life Expectancy vs. Vaccination Coverage')
plt.show()
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.scatter(LXA2['Hepatitis B'],LXA2['Life expectancy '])
plt.xlabel('Hepatitis B')
plt.ylabel('Life Expectancy')

plt.subplot(1,3,2)
plt.scatter(LXA2['Polio'],LXA2['Life expectancy '])
plt.xlabel('Polio')
plt.ylabel('Life Expectancy')
```

```python
correlation_matrix = LXA2[['infant deaths', 'under-five deaths ']].corr()
print(correlation_matrix)
```

```python
correlation_matrix = LXA2[['Life expectancy ','GDP','percentage expenditure']].corr()
print(correlation_matrix)
```

```python
correlation_expenditure = LXA2['Adult Mortality'].corr(LXA2['Total expenditure'])

print(f"Correlation between Adult Mortality and Total Expenditure: {correlation_expenditure}")

plt.figure(figsize=(7, 6))
sns.scatterplot(y='Total expenditure', x='Adult Mortality', data=LXA2)
plt.title('Impact of Total Expenditure on Adult Mortality')
plt.xlabel('Total Expenditure')
plt.ylabel('Adult Mortality')
plt.show()
```

```python
correlation_expenditure = LXA2['Life expectancy '].corr(LXA2['Total expenditure'])

print(f"Correlation between Life Expectancy and Total Expenditure: {correlation_expenditure}")

plt.figure(figsize=(7, 6))
sns.scatterplot(x='Total expenditure', y='Life expectancy ', data=LXA2)
plt.title('Impact of Total Expenditure on Life Expectancy')
plt.xlabel('Total Expenditure')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
correlation_expenditure = LXA2['Life expectancy '].corr(LXA2['Schooling'])

print(f"Correlation between Life Expectancy and schooling: {correlation_expenditure}")

plt.figure(figsize=(5,5))
sns.scatterplot(x='Schooling', y='Life expectancy ', data=LXA2)
plt.title('Impact of Schooling on Life Expectancy')
plt.xlabel('Schooling')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
correlation_expenditure = LXA2['Alcohol'].corr(LXA2['Adult Mortality'])

print(f"Correlation between Alcohol and Adult Mortality")

plt.figure(figsize=(5,5))
sns.scatterplot(x='Alcohol', y='Adult Mortality', data=LXA2)
plt.title('Impact of Alcohol consumption on Adult Mortality')
plt.xlabel('Alcohol')
plt.ylabel('Adult Mortality')
plt.show()
```

```python
avg_life_expectancy = LXA2.groupby('Country')['Life expectancy '].mean()
best_health_countries = avg_life_expectancy.nlargest(5)
worst_health_countries = avg_life_expectancy.nsmallest(5)

print("Countries with the Best Health Outcomes (Highest Average Life Expectancy):")
print(best_health_countries)

print("\nCountries with the Worst Health Outcomes (Lowest Average Life Expectancy):")
print(worst_health_countries)


import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.subplot(1, 2, 1)
best_health_countries.plot(kind='bar', color='brown')
plt.title('Top 5 Countries: Life Expectancy')
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy')

plt.subplot(1, 2, 2)
worst_health_countries.plot(kind='bar', color='salmon')
plt.title('Bottom 5 Countries: Life Expectancy')
plt.xlabel('Country')
plt.ylabel('Average Life Expectancy')

plt.tight_layout()
plt.show()
```

```python
correlation_population = LXA2['Life expectancy '].corr(LXA2['Population'])
print(f"Correlation between Life Expectancy and Population: {correlation_population}")

plt.figure(figsize=(5,5))
sns.scatterplot(x='Population', y='Life expectancy ', data=LXA2)
plt.title('Impact of Population Size on Life Expectancy')
plt.xlabel('Population Size')
plt.ylabel('Life Expectancy')
plt.show()
```
